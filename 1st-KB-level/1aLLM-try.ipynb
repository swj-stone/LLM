```python
# create_local_novel.py
import os

def create_local_novel_corpus():
    """纯本地生成小说语料，无需网络"""
    
    chapters = []
    
    # 生成10个不同的故事片段
    stories = [
        ("都市", "林晓", "陈明", "咖啡馆"),
        ("校园", "苏晴", "李华", "图书馆"),
        ("武侠", "林风", "白若雪", "竹林"),
        ("奇幻", "艾莉丝", "杰克", "魔法森林"),
        ("悬疑", "王警官", "神秘人", "老宅"),
    ]
    
    for style, name1, name2, place in stories:
        chapter = f"""
【{style}小说片段】

在{place}里，{name1}第一次遇见了{name2}。
那是一个雨后的午后，空气格外清新。
"{name2}，你也在这里？" {name1}有些惊讶。
{name2}转过身，露出微笑："真巧，我们又见面了。"

两人聊了很久，从天气聊到人生，从过去聊到未来。
{name1}发现，和{name2}在一起，时间总是过得特别快。

夕阳西下，{name2}说："天色不早了，我送你回去吧。"
{name1}点点头，心里涌起一股暖意。

就这样，一段故事开始了。
"""
        chapters.append(chapter)
    
    # 重复多次，增加数据量
    final_text = "\n\n".join(chapters * 100)  # 100遍，约200KB
    
    # 获取保存路径
    current_dir = os.getcwd()
    filename = 'local_novel_corpus.txt'
    full_path = os.path.join(current_dir, filename)
    
    with open(filename, 'w', encoding='utf-8') as f:
        f.write(final_text)
    
    size_kb = os.path.getsize(filename) / 1024
    print(f"✅ 生成成功!")
    print(f"   文件名: {filename}")
    print(f"   文件大小: {size_kb:.2f} KB ({size_kb/1024:.2f} MB)")
    print(f"   保存位置: {full_path}")
    print(f"   总字符数: {len(final_text):,}")
    
    return full_path

# 运行
if __name__ == "__main__":
    create_local_novel_corpus()
```

```python
# train_first_llm_fixed.py
import torch
import torch.nn as nn
from torch.nn import functional as F
import os

# ---------- 1. 准备数据 ----------
print("正在加载小说语料...")
file_path = 'local_novel_corpus.txt'  # 你刚才生成的文件

# 检查文件是否存在
if not os.path.exists(file_path):
    print(f"❌ 错误：找不到 {file_path} 文件！")
    print("请先运行 create_local_novel.py 生成语料文件")
    exit(1)

with open(file_path, 'r', encoding='utf-8') as f:
    text = f.read()

# 获取所有不重复的字符
chars = sorted(list(set(text)))
vocab_size = len(chars)  # 修正：删除了多余的右括号
print(f"语料总字符数: {len(text):,}")
print(f"词汇表大小（不同字符数）: {vocab_size}")

# 创建字符到数字的映射
stoi = {ch: i for i, ch in enumerate(chars)}
itos = {i: ch for i, ch in enumerate(chars)}
encode = lambda s: [stoi[c] for c in s]  # 编码器
decode = lambda l: ''.join([itos[i] for i in l])  # 解码器

# 将整个文本转换为张量
data = torch.tensor(encode(text), dtype=torch.long)
n = int(0.9 * len(data))
train_data = data[:n]
val_data = data[n:]

print(f"训练集大小: {len(train_data):,} 字符")
print(f"验证集大小: {len(val_data):,} 字符")

# ---------- 2. 定义超轻量级模型 ----------
class MiniLLM(nn.Module):
    def __init__(self, vocab_size, embedding_dim=128, hidden_dim=256, num_layers=2):
        super().__init__()
        self.num_layers = num_layers
        self.hidden_dim = hidden_dim
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x, hidden=None):
        x = self.embedding(x)
        out, hidden = self.lstm(x, hidden)
        out = self.fc(out)
        return out, hidden

    def init_hidden(self, batch_size, device='cpu'):
        return (torch.zeros(self.num_layers, batch_size, self.hidden_dim).to(device),
                torch.zeros(self.num_layers, batch_size, self.hidden_dim).to(device))

# 检测设备
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"\n使用设备: {device}")

# 初始化模型、损失函数、优化器
model = MiniLLM(vocab_size).to(device)
criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ---------- 3. 训练模型 ----------
print("\n开始训练...")
batch_size = 32  # 减小批次大小，适应小数据集
block_size = 64  # 减小序列长度
epochs = 30      # 减少轮次，避免过拟合

model.train()
for epoch in range(epochs):
    # 每个epoch开始前打乱数据
    idx = torch.randperm(len(train_data) - block_size)
    total_loss = 0
    steps = 0
    
    for i in range(0, len(idx), batch_size):
        batch_idx = idx[i:i+batch_size]
        if len(batch_idx) < batch_size:
            continue
            
        # 构建输入x和目标y
        x = torch.stack([train_data[i:i+block_size] for i in batch_idx]).to(device)
        y = torch.stack([train_data[i+1:i+block_size+1] for i in batch_idx]).to(device)
        
        # 前向传播
        hidden = model.init_hidden(batch_size, device)
        output, hidden = model(x, hidden)
        loss = criterion(output.view(-1, vocab_size), y.view(-1))
        
        # 反向传播
        optimizer.zero_grad()
        loss.backward()
        torch.nn.utils.clip_grad_norm_(model.parameters(), 1.0)  # 梯度裁剪
        optimizer.step()
        
        total_loss += loss.item()
        steps += 1
        
        if steps % 20 == 0:
            print(f"Epoch {epoch+1}/{epochs}, Step {steps}, Loss: {loss.item():.4f}")
    
    avg_loss = total_loss / steps
    print(f"📊 Epoch {epoch+1}/{epochs} 完成，平均损失: {avg_loss:.4f}")

print("✅ 训练完成！")

# ---------- 4. 生成小说 ----------
print("\n🎭 生成小说...")
model.eval()

# 多个起始提示
prompts = ["【都市", "【武侠", "【校园", "在咖啡馆"]
start_prompt = prompts[0]  # 可以换成其他的

input_ids = torch.tensor([encode(start_prompt)]).to(device)
hidden = model.init_hidden(1, device)

generated_ids = []
with torch.no_grad():
    for _ in range(200):  # 生成200个字符
        output, hidden = model(input_ids, hidden)
        logits = output[0, -1, :] / 0.8  # 温度参数
        probs = F.softmax(logits, dim=-1)
        next_char_idx = torch.multinomial(probs, 1).item()
        generated_ids.append(next_char_idx)
        input_ids = torch.tensor([[next_char_idx]]).to(device)

generated_text = decode(generated_ids)
print("\n" + "="*50)
print("生成的文本如下：")
print("="*50)
print(start_prompt + generated_text)
print("="*50)

# ---------- 5. 保存模型（可选）----------
save_model = input("\n是否保存模型？(y/n): ").lower()
if save_model == 'y':
    torch.save(model.state_dict(), 'my_first_llm.pth')
    print("✅ 模型已保存为 my_first_llm.pth")
```

```python
# use_model.py - 单独使用训练好的模型生成小说
import torch
import torch.nn as nn
from torch.nn import functional as F
import os

# ---------- 1. 重新定义相同的模型结构 ----------
class MiniLLM(nn.Module):
    def __init__(self, vocab_size, embedding_dim=128, hidden_dim=256, num_layers=2):
        super().__init__()
        self.num_layers = num_layers
        self.hidden_dim = hidden_dim
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.lstm = nn.LSTM(embedding_dim, hidden_dim, num_layers, batch_first=True)
        self.fc = nn.Linear(hidden_dim, vocab_size)
    
    def forward(self, x, hidden=None):
        x = self.embedding(x)
        out, hidden = self.lstm(x, hidden)
        out = self.fc(out)
        return out, hidden
    
    def init_hidden(self, batch_size, device='cpu'):
        return (torch.zeros(self.num_layers, batch_size, self.hidden_dim).to(device),
                torch.zeros(self.num_layers, batch_size, self.hidden_dim).to(device))

# ---------- 2. 重建词汇表（关键！）----------
# 需要从原始语料中重建相同的映射关系
def rebuild_vocab_from_corpus(corpus_path='local_novel_corpus.txt'):
    """从语料文件重建词汇表"""
    if not os.path.exists(corpus_path):
        print(f"❌ 找不到语料文件: {corpus_path}")
        print("请确保 local_novel_corpus.txt 在当前目录")
        return None, None, None
    
    with open(corpus_path, 'r', encoding='utf-8') as f:
        text = f.read()
    
    chars = sorted(list(set(text)))
    vocab_size = len(chars)
    stoi = {ch: i for i, ch in enumerate(chars)}
    itos = {i: ch for i, ch in enumerate(chars)}
    
    print(f"✅ 重建词汇表完成，大小: {vocab_size}")
    return vocab_size, stoi, itos

# 重建词汇表
vocab_size, stoi, itos = rebuild_vocab_from_corpus('local_novel_corpus.txt')

if vocab_size is None:
    exit(1)

# 编码/解码函数
encode = lambda s: [stoi[c] for c in s]
decode = lambda l: ''.join([itos[i] for i in l])

# ---------- 3. 加载模型 ----------
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"使用设备: {device}")

# 创建模型
model = MiniLLM(vocab_size=vocab_size).to(device)

# 加载权重
model_path = 'my_first_llm.pth'
if not os.path.exists(model_path):
    print(f"❌ 找不到模型文件: {model_path}")
    print("请确保 my_first_llm.pth 在当前目录")
    exit(1)

model.load_state_dict(torch.load(model_path, map_location=device))
model.eval()
print(f"✅ 模型加载成功: {model_path}")

# ---------- 4. 生成小说函数 ----------
def generate(prompt, max_length=300, temperature=0.8, top_k=50):
    """
    生成小说续写
    prompt: 起始提示词
    max_length: 最大生成字符数
    temperature: 温度（越高越随机，越低越保守）
    top_k: 只从概率最高的k个字符中采样
    """
    model.eval()
    
    # 编码提示词
    input_ids = torch.tensor([encode(prompt)]).to(device)
    hidden = model.init_hidden(1, device)
    
    generated_ids = []
    
    with torch.no_grad():
        for _ in range(max_length):
            # 前向传播
            output, hidden = model(input_ids, hidden)
            
            # 获取最后一个字符的logits
            logits = output[0, -1, :] / temperature
            
            # Top-k采样
            probs = F.softmax(logits, dim=-1)
            probs_top, indices = torch.topk(probs, top_k)
            next_char_idx = indices[torch.multinomial(probs_top, 1)].item()
            
            generated_ids.append(next_char_idx)
            input_ids = torch.tensor([[next_char_idx]]).to(device)
    
    # 解码并返回
    generated_text = decode(generated_ids)
    return prompt + generated_text

# ---------- 交互式生成（简化版）---------
print("\n" + "="*60)
print("小说生成器")
print("="*60)
print("输入小说开头，输入 'quit' 或 'q' 或 '退出' 退出程序")
print("示例开头：咖啡馆、【武侠、两个人、神秘")
print("-"*60)

while True:
    # 只询问开头
    user_input = input("\n📝 请输入小说开头: ").strip()
    
    # 退出判断
    if user_input.lower() in ['quit', 'q', '退出']:
        print("\n👋 再见！")
        break
    
    # 检查是否为空
    if len(user_input) == 0:
        print("⚠️ 请输入一些文字作为开头")
        continue
    
    # 检查是否包含未知字符
    unknown_chars = [c for c in user_input if c not in stoi]
    if unknown_chars:
        print(f"⚠️ 警告：以下字符不在词汇表中: {''.join(unknown_chars)}")
        print("   可能影响生成效果，建议使用语料中的字")
    
    # 直接生成，不询问其他参数（使用默认值：最大长度200，温度0.8）
    print("\n🤖 生成中...\n")
    try:
        result = generate(user_input, max_length=200, temperature=0.8)
        print("-" * 40)
        print(result)
        print("-" * 40)
    except KeyError as e:
        print(f"❌ 错误：字符 '{e.args[0]}' 不在词汇表中")
        print("   请换一个开头试试")
```
